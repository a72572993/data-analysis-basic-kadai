```python
#1.カテゴリー列の要素(弁当、デザートなど)の出現頻度をカウントして、その結果を棒グラフで表示する

#Step1.必要なライブラリをインポートする
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

#csvファイルをデータフレームとして読み込む
df = pd.read_csv("sample_pandas_6.csv")

category_df = pd.read_csv('category.csv')
category_df

#Step2.カテゴリー列の要素(弁当・デザートなど)のユニークな出現頻度をカウントするメソッドと、そのメソッドの実行結果を棒グラフ化するメソッドを組み合わせたコードを記述する。

#商品番号で結合させる
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

#カテゴリー列の要素のカウント
category_counts = df['カテゴリー'].value_counts()
category_counts
#棒グラフ化
category_counts.plot(kind='bar')
plt.title("カテゴリー列の要素出題頻度")

#Step3.グラフを表示させる行「plt.show()」を書く
plt.show()

#2.商品番号列の商品番号ごとに、注文数の基本統計量を求める

#Step4.ある基準列(商品番号列)ごとに要素をグループにまとめサブ列を集計するメソッドと、基本統計量を算出するメソッドを組み合わせたコードを記載する
#データフレームを基準列ごとに要素をグループにまとめ、サブ列を集計するメソッド
#> df.groupby('商品番号')['注文数'].集計関数()
#データフレームの基本統計量を算出するメソッド
df.groupby('商品番号')['注文数'].describe()



```
